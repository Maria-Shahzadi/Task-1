```python
from collections import defaultdict

# Generator function (stream file line by line)
def stream_errors(file_path):
    with open(file_path, "r") as file:
        for line in file:
            if "ERROR" in line:   # only critical errors
                yield line.strip()


# Function to group errors and show live count
def process_errors(file_path):
    error_counts = defaultdict(int)

    print("\n===== LIVE ERROR STREAM =====\n")

    for error in stream_errors(file_path):

        # Example error format: ERROR: DatabaseFailure
        parts = error.split(":")
        if len(parts) > 1:
            error_type = parts[1].strip()
        else:
            error_type = "Unknown"

        error_counts[error_type] += 1

        print(f"Error Type: {error_type} | Count: {error_counts[error_type]}")


    print("\n===== FINAL ERROR SUMMARY =====\n")
    for err, count in error_counts.items():
        print(f"{err}  →  {count}")
        

# Example log file creation (for testing in Jupyter)
with open("server_log.txt", "w") as f:
    f.write("""INFO: Server started
ERROR: DatabaseFailure
WARNING: Disk space low
ERROR: TimeoutError
INFO: User login
ERROR: DatabaseFailure
ERROR: TimeoutError
ERROR: AuthenticationError
""")


# Run the system
process_errors("server_log.txt")
```
